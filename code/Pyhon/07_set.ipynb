```python
# set ì—°ì‚°
# ì§‘í•©ì˜ ì—°ì‚° : í•©ì§‘í•©, êµì§‘í•©, ì°¨ì§‘í•©, ëŒ€ì¹­ì°¨ì§‘í•©
a = {1,2,3}
b = {3,4,5}

# í•©ì§‘í•©(|, .union)
s_union1 = a | b
s_union2 = a.union(b)
print("í•©ì§‘í•©1", s_union1)
print("í•©ì§‘í•©2", s_union2)

# êµì§‘í•©(&, .intersection)
s_inter1 = a & b
s_inter2 = a.intersection(b)
print("êµì§‘í•©1", s_inter1)
print("êµì§‘í•©2", s_inter2)

# ì°¨ì§‘í•©
s_diff1 = a -b
s_diff2 = a.difference(b)
print("ì°¨ì§‘í•©1", s_diff1)
print("ì°¨ì§‘í•©2", s_diff2)

# ëŒ€ì¹­ì°¨ì§‘í•©
s_symm1 = (a | b) - (a & b)
s_symm2 = a ^ b
s_symm3 = a.symmetric_difference(b)

print("ëŒ€ì¹­ì°¨ì§‘í•©1", s_symm1)
print("ëŒ€ì¹­ì°¨ì§‘í•©2", s_symm2)
print("ëŒ€ì¹­ì°¨ì§‘í•©3", s_symm3)
```

```python
# set ë©”ì„œë“œ
s1 = {1, 2, 3}

# ì›ì†Œ ì¶”ê°€
s1.add(4)
print("ì›ì†Œ ì¶”ê°€", s1)

# ì—¬ëŸ¬ ì›ì†Œ ì¶”ê°€
s1.update((5,6,7))
print("ì—¬ëŸ¬ ì›ì†Œ ì¶”ê°€", s1)

# ì›ì†Œ ì œê±°
s1.remove(4)
print("ì›ì†Œ ì œê±°1", s1)
# s1.remove(100) # ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ì›ì†Œ ì‚­ì œì‹œë„ì‹œ ì—ëŸ¬
s1.discard(100)
s1.discard(6)
print("ì›ì†Œ ì œê±°2", s1)
deleted = s1.pop() # ì„ì˜ì˜ ê°’ í•˜ë‚˜ ì œê±°
print("ì›ì†Œ ì œê±°3", s1, deleted)
```

```python
# ë¶€ë¶„ì§‘í•©(subset) ê´€ë ¨ ë©”ì„œë“œ
a = {10, 20, 30, 40, 50} # ìƒìœ„ì§‘í•©(superset)
b = {20, 30, 40} # ë¶€ë¶„ì§‘í•©(subset)
c = {10, 200, 300, 400, 500}

# ë¶€ë¶„ì§‘í•© ì—¬ë¶€ íŒë‹¨
print(b.issubset(a))
print(a.issubset(b))

# ìƒìœ„ì§‘í•© ì—¬ë¶€ íŒë‹¨
print(a.issuperset(b))
print(b.issuperset(a))

# ê³µí†µ ì›ì†Œê°€ ì—†ëŠ”ì§€ í™•ì¸
print(a.isdisjoint(b))
print(a.isdisjoint(c))
print(b.isdisjoint(c))
```

```python
# ë¬¸ì œ 1. ì¤‘ë³µ ì œê±° ë° ê°œìˆ˜ ì„¸ê¸° 
# ì–´ë–¤ í•™ê¸‰ì˜ í•™ìƒë“¤ì´ ì œì¶œí•œ íŒ€ ê³¼ì œ íŒŒì¼ ì´ë¦„ ëª©ë¡ì´ ì•„ë˜ì™€ ê°™ìŠµë‹ˆë‹¤.
# ì¤‘ë³µ ì œì¶œëœ ê²½ìš°ë„ í¬í•¨ë˜ì–´ ìˆìŠµë‹ˆë‹¤.
# ì¤‘ë³µì„ ì œê±°í•œ í›„, ì´ ëª‡ ëª…ì´ ì œì¶œí–ˆëŠ”ì§€ ì¶œë ¥í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.
submissions = ['kim', 'lee', 'kim', 'park', 'choi', 'lee', 'lee']

unique_studnets = set(submissions)

# í•™ìƒìˆ˜
print(f"ì œì¶œí•œ í•™ìƒìˆ˜: {len(unique_studnets)}") 
# ì œì¶œìëª…ë‹¨
print(f"ì œì¶œì ëª…ë‹¨: {unique_studnets}")
```

```python
# ë¬¸ì œ 2. ê³µí†µ ê´€ì‹¬ì‚¬ ì°¾ê¸°
# ë‘ ëª…ì˜ ì‚¬ìš©ìê°€ ê°ì ì¢‹ì•„í•˜ëŠ” ì˜í™” ì¥ë¥´ë¥¼ ì•„ë˜ì™€ ê°™ì´ ì…ë ¥í–ˆìŠµë‹ˆë‹¤.
# ë‘ ì‚¬ìš©ìì˜ ê³µí†µ ê´€ì‹¬ ì¥ë¥´, ì„œë¡œ ë‹¤ë¥¸ ì¥ë¥´, ëª¨ë“  ì¥ë¥´ ëª©ë¡ì„ ì¶œë ¥í•˜ì„¸ìš”.
user1 = {'SF', 'Action', 'Drama'}
user2 = {'Drama', 'Romance', 'Action'}

# ê³µí†µ ê´€ì‹¬ ì¥ë¥´(êµì§‘í•©)
common = user1 & user2

# ì„œë¡œ ë‹¤ë¥¸ ì¥ë¥´(ëŒ€ì¹­ì°¨ì§‘í•©)
diff = user1 ^ user2

# ëª¨ë“  ì¥ë¥´ ëª©ë¡(í•©ì§‘í•©)
total = user1 | user2

print(f"ê³µí†µ ê´€ì‹¬ ì¥ë¥´: {common}")
print(f"ì„œë¡œ ë‹¤ë¥¸ ì¥ë¥´: {diff}")
print(f"ëª¨ë“  ì¥ë¥´ ëª©ë¡: {total}")
```

```python
# ë¬¸ì œ 3. ë¶€ë¶„ì§‘í•© ê´€ê³„ íŒë‹¨
# ì–´ë–¤ ìœ ì €ê°€ ê°€ì§€ê³  ìˆëŠ” ìê²©ì¦ ëª©ë¡ê³¼ íŠ¹ì • ì§ë¬´ì— í•„ìš”í•œ ìê²©ì¦ ëª©ë¡ì´ ì£¼ì–´ì§‘ë‹ˆë‹¤.
# ì´ ì‚¬ìš©ìê°€ ì§€ì› ìê²©ì„ ê°–ì¶”ì—ˆëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.
my_certificates = {'SQL', 'Python', 'Linux'}
job_required = {'SQL', 'Python'}

# ì§€ì› ìê²© ì¶©ì¡± ì—¬ë¶€
is_qualified = job_required.issubset(my_certificates)

print(f"ì§€ì› ìê²© ì¶©ì¡± ì—¬ë¶€: {is_qualified}")
```
