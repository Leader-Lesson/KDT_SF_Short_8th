```python
# ì‹¤ìŠµ 1.
# í•¨ìˆ˜ ì´ë¦„ì€ calculateë¡œ í•©ë‹ˆë‹¤.
# ë§¤ê°œë³€ìˆ˜ëŠ” a, b, operator ì„¸ ê°œì…ë‹ˆë‹¤.
# operatorëŠ” ë¬¸ìì—´ì´ë©°, ë‹¤ìŒ ì¤‘ í•˜ë‚˜ì…ë‹ˆë‹¤: "+", "-", "*", "/"
# ë‚˜ëˆ—ì…ˆì€ ê²°ê³¼ë¥¼ ì‹¤ìˆ˜(float) ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.
# ì˜¬ë°”ë¥´ì§€ ì•Šì€ ì—°ì‚°ìê°€ ë“¤ì–´ì˜¤ë©´ "ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°ì…ë‹ˆë‹¤"ë¼ëŠ” ë¬¸ìì—´ì„ ë°˜í™˜í•˜ì„¸ìš”.

def calculate(a, b, operator):
  if operator == "+":
    return a + b
  elif operator == "-":
    return a - b
  elif operator == "*":
    return a * b
  elif operator == "/":
    return float(a / b)
  else:
    return "ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°ì…ë‹ˆë‹¤"
  
print(calculate(20,30,"+"))
print(calculate(50,40,"-"))
print(calculate(5,4,"*"))
print(calculate(100,3,"/"))
print(calculate(1000,10000,"&"))

print(calculate(30,50,"+") + calculate(10,20,"+"))
```

```python
# ì‹¤ìŠµ 3. ì „ì—­ ë³€ìˆ˜ ì—°ìŠµí•˜ê¸°
# âœ… ìš”êµ¬ì‚¬í•­
# ì „ì—­ ë³€ìˆ˜ current_userëŠ” ë¡œê·¸ì¸í•œ ì‚¬ìš©ìì˜ ì´ë¦„ì„ ì €ì¥í•©ë‹ˆë‹¤.
# login(name) í•¨ìˆ˜ëŠ” ì‚¬ìš©ìë¥¼ ë¡œê·¸ì¸ì‹œí‚¤ê³ , logout() í•¨ìˆ˜ëŠ” ë¡œê·¸ì•„ì›ƒ ìƒíƒœë¡œ ë§Œë“­ë‹ˆë‹¤.
# ì´ë¯¸ ë¡œê·¸ì¸ëœ ìƒíƒœì—ì„œ ë‹¤ì‹œ ë¡œê·¸ì¸í•˜ë©´ "ì´ë¯¸ ë¡œê·¸ì¸ë˜ì–´ ìˆìŠµë‹ˆë‹¤"ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
# ë¡œê·¸ì•„ì›ƒí•˜ì§€ ì•Šê³  ë¡œê·¸ì¸ì„ ì—¬ëŸ¬ ë²ˆ ì‹œë„í•  ìˆ˜ ì—†ë„ë¡ í•©ë‹ˆë‹¤.

current_user = None
login_count = 0

def login(name):
  global current_user
  global login_count

  if current_user == None:
    if len(name) > 4:
      current_user = name
      print(f"ğŸ¤—{name}ë‹˜ ë¡œê·¸ì¸ ì„±ê³µ!")
    # ì˜ˆì™¸ì²˜ë¦¬
    else:
      print("âš ï¸ì•„ì´ë””ëŠ” ë„¤ê¸€ì ì´ìƒì´ì–´ì•¼ í•´ìš”.")
      
      login_count += 1
      if login_count > 4:
        print("ë”ì´ìƒ ë¡œê·¸ì¸ ì‹œë„ë¥¼ í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
  else:
    print("ğŸš¨ì´ë¯¸ ë¡œê·¸ì¸ë˜ì–´ ìˆìŠµë‹ˆë‹¤.")

    login_count += 1
    if login_count > 4:
      print("ë”ì´ìƒ ë¡œê·¸ì¸ ì‹œë„ë¥¼ í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")


def logout():
  global current_user
  global login_count 
  if current_user == None:
    print("âš ï¸ë¡œê·¸ì¸ ìƒíƒœê°€ ì•„ë‹™ë‹ˆë‹¤.")
  else:
    print("âœ…ë¡œê·¸ì•„ì›ƒ ë˜ì—ˆìŠµë‹ˆë‹¤!")
    current_user = None
    login_count = 0

login("")
login("a")
login("b")
login("c")
login("de")
```

```python
import time
# ì¬ê·€í•¨ìˆ˜
# 1. ìê¸°ê°€ ìê¸° ìì‹ ì„ í˜¸ì¶œí•˜ëŠ” í•¨ìˆ˜
# 2. ë°˜ë“œì‹œ ê¸°ë³¸ ì¡°ê±´(ì¢…ë£Œ ì¡°ê±´)ì´ ìˆì–´ì•¼ í•¨
# - í° ë¬¸ì œë¥¼ ì‘ì€ ë¬¸ì œë¡œ ë‚˜ëˆ„ì—ˆì„ ë•Œ ì¼ì •í•œ íŒ¨í„´ì´ ìˆì–´ì•¼ í•¨

def recursive_func(n):
  # ê¸°ë³¸ ì¡°ê±´
  if n == 0:
    return
  recursive_func(n-1)
  print("ì¬ê·€ í˜¸ì¶œ", n)
  time.sleep(1)

recursive_func(5)
```

```python
# ì‹¤ìŠµ 4. ê±°ë“­ ì œê³±
# ë°˜ë³µë¬¸ìœ¼ë¡œ êµ¬í˜„
def power_for(a, n):
  result = 1
  for _ in range(n):
    result *= a
  return result

print(power_for(4,3))

# ì¬ê·€í•¨ìˆ˜ êµ¬í˜„
def power_rec(a, n):
  if n == 0:
    return 1
  return a * power_rec(a, n - 1)

print("ì¬ê·€í•¨ìˆ˜", power_rec(4,3))
```

```python
# ì‹¤ìŠµ 5. íŒ©í† ë¦¬ì–¼
# ë°˜ë³µë¬¸ìœ¼ë¡œ êµ¬í˜„
def factorial_for(n):
  # ì˜ˆì™¸ì²˜ë¦¬
  if n < 0:
    return "ìŒìˆ˜ì˜ íŒ©í† ë¦¬ì–¼ì€ ì •ì˜ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤."
  
  result = 1

  for i in range(1, n+1):
    result *= i
  
  return result

factorial_for(5)

# ì¬ê·€í•¨ìˆ˜
def factorial_rec(n):
  # ì˜ˆì™¸ì²˜ë¦¬
  if n < 0:
    return "ìŒìˆ˜ì˜ íŒ©í† ë¦¬ì–¼ì€ ì •ì˜ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤."
  
  # ê¸°ë³¸ì¡°ê±´
  if n == 0 or n == 1:
    return 1
  
  return n * factorial_rec(n - 1)

factorial_rec(-10)
factorial_rec(0)
factorial_rec(5)
```

```python
# ì‹¤ìŠµ 6. í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´
# ë°˜ë³µë¬¸
def fibonacci_for(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  
  a,b = 0,1
  for _ in range(n-1):
    a,b = b,a+b

  return b

fibonacci_for(6)

# ì¬ê·€í•¨ìˆ˜
def fibonacci_rec(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  
  return fibonacci_rec(n-1) + fibonacci_rec(n-2)

fibonacci_rec(20)
```
