```python
class Animal:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    
  def bark(self):
    print("ë™ë¬¼ì´ ìš¸ìŒ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.")

# ìžë…€ í´ëž˜ìŠ¤
class Dog(Animal):
  # ì˜¤ë²„ë¼ì´ë”©
  def __init__(self, name, age, species):
    # superëŠ” ë¶€ëª¨ë¥¼ ê°€ë¦¬í‚´
    # ë¶€ëª¨ì˜ í•¨ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ ë³µì œí•¨
    super().__init__(name, age)
    self.species = species

  # ì˜¤ë²„ë¼ì´ë”©
  def bark(self):
    super().bark()
    print("ë©ë©")

d = Dog("ë¦¬ì¹˜", 12, "í¬ë©”ë¼ë‹ˆì•ˆ")
d.bark()
print(d.name)
print(d.age)
print(d.species)
```

```python
# ì¶”ìƒ í´ëž˜ìŠ¤(Abstract Class)
# í´ëž˜ìŠ¤ì˜ êµ¬ì¡°ë¥¼ ì •ì˜í•˜ëŠ” í´ëž˜ìŠ¤

from abc import ABC, abstractmethod

class Animal(ABC):
  @abstractmethod
  def bark(self):
    pass

class Dog(Animal):
  def bark(self):
    print("ë©ë©")

# a = Animal()
d = Dog()
d.bark()
```

```python
# ì‹¤ìŠµ5. ì¶”ìƒ í´ëž˜ìŠ¤ Payment êµ¬í˜„
from abc import ABC, abstractmethod

class Payment(ABC):
  @abstractmethod
  def pay(self, amount):
    pass

class CardPayment(Payment):
  def __init__(self):
    super().__init__()

  def pay(self, amount):
    print(f"ì¹´ë“œë¡œ {amount}ì›ì„ ê²°ì œí•©ë‹ˆë‹¤.")

class CashPayment(Payment):
  def pay(self, amount):
    print(f"í˜„ê¸ˆìœ¼ë¡œ {amount}ì›ì„ ê²°ì œí•©ë‹ˆë‹¤.")

card = CardPayment()
card.pay(10000)
```
