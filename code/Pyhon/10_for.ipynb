```python
# ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜(List Comprehension)
# forë¬¸ì„ ë¦¬ìŠ¤íŠ¸ì— í•œì¤„ë¡œ ì¶•ì•½í•˜ì—¬ ìƒˆë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•˜ëŠ” ë¬¸ë²•
# [í‘œí˜„ì‹(ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œ) for ë³€ìˆ˜ in ë°˜ë³µëŒ€ìƒ if ì¡°ê±´]
# í‘œí˜„ì‹ : ê°’ì„ ìœ ë„í•˜ëŠ” ì‹(í‘œí˜„)

# forë¬¸ ì´ìš©
squares = []
for x in range(1,6):
  squares.append(x ** 2)
print(squares)

# ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜
squares_2 = [x ** 2 for x in range(1,6)]
print(squares_2)
```
